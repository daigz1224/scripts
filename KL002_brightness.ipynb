```python
from PIL import Image, ImageStat
import os
import matplotlib.pyplot as plt
```

```python
def cal_brightness(dirpath):
    dirname = dirpath.split("/")[-2]
    jpgs = os.listdir(dirpath)
    index = list(range(1, len(jpgs) + 1, 1))
    bris = []
    for jpg in jpgs:
        path = dirpath + jpg
        im = Image.open(path)
        stat = ImageStat.Stat(im)
        bri = stat.sum[0] / stat.count[0]
        bris.append(bri)
    plt.figure()
    plt.scatter(index, bris, s=1)
    plt.xlabel("frame(s)")
    plt.ylabel("brightness")
    plt.title(dirname + ': ' + str(len(jpgs)) + ' frames in total')
```
